```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)    #line plotplt
plt.show()
```

```python
plt.plot(x,y)  
plt.title("first line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle=':',linewidth=3)  
plt.title("first line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='orange',linewidth=3)  
plt.plot(x,y2,color='green',linewidth=2)  
plt.grid(True)
plt.title("first line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.subplot(1,2,1)           #along row
plt.plot(x,y1,color='yellow',linestyle=':',linewidth=3)

plt.subplot(1,2,2)
plt.plot(x,y2,color='red',linestyle=':',linewidth=2)

plt.show()
```

```python
#along columns
plt.subplot(2,1,1)
plt.plot(x,y1,color='yellow',linestyle=':',linewidth=3)

plt.subplot(2,1,2)
plt.plot(x,y2,color='red',linestyle=':',linewidth=2)

plt.show()
```

```python
#Bar Plot

student= {'aman':34,"mat":56,'sam':8989}

names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title("marks of student")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
#horizonatl bar plot
plt.barh(names,values,color='green')
plt.title("marks of student")
plt.xlabel("marks")
plt.ylabel("names")
plt.grid(True)
plt.show()
```

```python
#scatter plot

x=[1,3,5,6,7,8]
y=[6,7,8,9,2,1]

plt.scatter(x,y)
plt.show()
```

```python
x=[1,3,5,6,7,8]
y=[6,7,8,9,2,1]

plt.scatter(x,y,marker="*",c="purple",s=340)
plt.show()
```

```python
x=[1,3,5,6,7,8]
y=[6,7,8,9,2,1]
z=[3,6,4,7,8,4]

plt.scatter(x,y,marker="*",c="purple",s=340)
plt.scatter(x,z,c="red",s=200)
plt.show()
```

```python
x=[1,3,5,6,7,8]
y=[6,7,8,9,2,1]
z=[3,6,4,7,8,4]

plt.subplot(1,2,1)
plt.scatter(x,y,marker="*",c="purple",s=340)

plt.subplot(1,2,2)
plt.scatter(x,z,c="red",s=200)
plt.show()
```

```python
x=[1,3,5,6,7,8]
y=[6,7,8,9,2,1]
z=[3,6,4,7,8,4]

plt.subplot(2,1,1)
plt.scatter(x,y,marker="*",c="purple",s=340)

plt.subplot(2,1,2)
plt.scatter(x,z,c="red",s=200)
plt.show()
```

```python
#histogram

li=[1,2,2,4,3,5,4,6,7,8]
plt.hist(li)
plt.show()
```

```python
plt.hist(li,color='g',bins=3)
plt.show()
```

```python
#box plot
x=[1,3,5,6,7,8]
y=[6,7,8,9,5,4]
z=[3,6,4,7,8,4]

ans=list([x,y,z])
```

```python
plt.boxplot(ans)
plt.show()
```

```python
#voilen plot
x=[1,3,5,6,7,8]
y=[6,7,8,9,5,4]
z=[3,6,4,7,8,4]

ans=list([x,y,z])
```

```python
plt.violinplot(ans,showmedians=True)
plt.show()
```

```python
#pie chart
fruit=['apple','mango','banana','litchi']
quanti=[34,56,87,55]
plt.pie(quanti,labels=fruit)
plt.show()
```

```python
fruit=['apple','mango','banana','litchi']
quanti=[34,56,87,55]
plt.pie(quanti,labels=fruit ,autopct="%0.1f%%",colors=['red','green','orange','blue'])
plt.show()
```

```python
fruit=['apple','mango','banana','litchi']
quanti=[34,56,87,55]
plt.pie(quanti,labels=fruit ,autopct="%0.1f%%",colors=['red','green','orange','blue'],radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
